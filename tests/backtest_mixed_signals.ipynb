```python
symbol = "TQQQ"
start = "2025-03-07"
end = "2025-03-28"

strategy = MixedSignalsStrategy(symbol)  # or MixedSignalsStrategy(symbol)
data = load_data(symbol, start, end)

data.head()
```

```python
signals = strategy.generate_signals(data)
signals.tail(10)
```

```python
print(signals['signal'].value_counts())
signals[['close', 'signal']].head(20)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# --- Setup ---
position = 0
cash = 240
starting_cash = cash
shares = 0
portfolio_values = []
buy_dates, sell_dates = [], []
buy_prices, sell_prices = [], []
trade_counts = {}

# --- Backtest Loop ---
for i in range(1, len(signals)):
    timestamp = signals.index[i]
    date = timestamp.date()
    signal = signals['signal'].iloc[i]
    price = signals['close'].iloc[i]

    # Init counter for the day
    if date not in trade_counts:
        trade_counts[date] = 0

    # Skip if already made 3 trades today
    if trade_counts[date] >= 5:
        portfolio_values.append(cash + shares * price)
        continue

    # Buy logic
    if signal == 1 and position == 0:
        shares = cash // price
        cash -= shares * price
        position = 1
        buy_dates.append(timestamp)
        buy_prices.append(price)
        trade_counts[date] += 1

    # Sell logic
    elif signal == -1 and position == 1:
        cash += shares * price
        shares = 0
        position = 0
        sell_dates.append(timestamp)
        sell_prices.append(price)
        trade_counts[date] += 1

    # Track portfolio value
    portfolio_values.append(cash + shares * price)

# --- Final Stats ---
final_value = cash + shares * signals['close'].iloc[-1]
return_pct = ((final_value - starting_cash) / starting_cash) * 100
print(f"ðŸ“Š Final portfolio value: ${final_value:.2f}")
print(f"ðŸ“ˆ Return: {return_pct:.2f}%")

# --- Plotting ---
timestamps = signals.index[1:]
portfolio_series = pd.Series(portfolio_values, index=timestamps)
price_series = signals['close'].iloc[1:]

plt.figure(figsize=(14, 7))
plt.plot(price_series.index, price_series.values, label="Stock Price (close)", alpha=0.5, color='gray')
plt.scatter(buy_dates, buy_prices, marker='^', color='green', label='Buy', s=80)
plt.scatter(sell_dates, sell_prices, marker='v', color='red', label='Sell', s=80)
plt.plot(portfolio_series.index, portfolio_series.values, label="Portfolio Value", color='blue', linewidth=2)
plt.title("Portfolio Value vs Stock Price (Max 3 Trades/Day)")
plt.xlabel("Time")
plt.ylabel("USD")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
